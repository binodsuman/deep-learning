```python
from keras.models import Sequential
from keras.layers import Conv2D, Dense
from IPython.display import Image
```

```python
Image(filename='ANN.jpg')
```

```python
model = Sequential()
# specify the input size for in the first layer.
model.add(Dense(5, activation='relu', input_shape= (3,)))
# Output layer
model.add(Dense(2, activation='softmax'))
model.summary()
```

```python
Image(filename='ANN2.jpg')
```

```python
model = Sequential()
# specify the input size for in the first layer.
model.add(Dense(100, activation='relu', input_shape= (50,)))
model.add(Dense(1, activation='relu'))
model.add(Dense(100, activation='relu'))
# Output layer
model.add(Dense(50, activation='softmax'))
model.summary()
```

```python
Image(filename='CNN.jpg')
```

```python
# total_params = (filter_height * filter_width * input_image_channels + 1) * number_of_filters
# use_bias=False then no add 1. By default it is True

model = Sequential()
model.add(Conv2D(1, (3,3),  input_shape=(5, 5, 1)))
model.summary()
```

```python
# total_params = (filter_height * filter_width * input_image_channels + 1) * number_of_filters
# use_bias=False then no add 1. By default it is True

model = Sequential()
model.add(Conv2D(1, (3,3),  input_shape=(50, 50, 1)))
model.summary()
```

```python
# Independent of input size
model = Sequential()
model.add(Conv2D(1, (3,3),  input_shape=(15, 15, 1)))
model.summary()
```

```python
Image(filename='CNN2.jpg')
```

```python
# total_params = (filter_height * filter_width * input_image_channels + 1) * number_of_filters
# use_bias=False then no add 1. By default it is True
model = Sequential()
model.add(Conv2D(1, (3,3),  input_shape=(5, 5, 1), use_bias=False))
model.summary()
```

```python
# LSTM
from keras.layers import LSTM

model = Sequential()
model.add(LSTM(units = 2, input_dim=3, input_length=16))
model.summary()


```
